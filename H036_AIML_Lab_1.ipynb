1. Hyperbolic Tangent Function
2.Logistic Function
3. Linear Function


```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp as e
```

```python
#for logistic function
x=[]
y=[]
for i in range(-20,20,1):
  a=1/(1+exp(-i))
  x.append(i)
  y.append(a)

plt.plot(x,y,color='black')
```

```python
#for linear function
x=[]
y=[]
for i in range(-20,20,1):
  x.append(i)
  y.append(i)

plt.plot(x,y,color='black')
```

```python
#for hyperbolic tangent function
x=[]
y=[]
for i in range(-20,20,1):
  a=exp(-i)
  b=exp(i)
  p=(b-a)/(b+a)
  x.append(i)
  y.append(p)

plt.plot(x,y,color='black')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

def return_exponent(number):
  return np.exp(number)

def return_negative_exponent(number):
  return np.exp(-number)

x_axis_values=np.linspace(start = -10, stop = 10, num = 201)
y_axis_values_logistic=[]
y_axis_values_tanh=[]

for x_value in x_axis_values:
  '''using the function return_negative_exponent to calculate the logistic
  function value according to formula'''

  logistic_value=1/(1+return_negative_exponent(x_value))
  #appending the value to list for plotting
  y_axis_values_logistic.append(logistic_value)

  '''using the function return_negative_exponent and return_nexponent
  to calculate the tanh function value according to formula'''

  numerator=return_exponent(x_value)-return_negative_exponent(x_value)
  denominator=return_exponent(x_value)+return_negative_exponent(x_value)
  tanh_value=(numerator)/(denominator)
  #appending the value to list for plotting
  y_axis_values_tanh.append(tanh_value)


plt.figure(1)
#plt.subplot(311)
plt.title('Linear Function ')
plt.plot(x_axis_values,x_axis_values,color='black')

#plt.subplot(312)
plt.figure(2)
plt.title('Logistic Function ')
plt.plot(x_axis_values,y_axis_values_logistic,color='black')

#plt.subplot(313)
plt.figure(3)
plt.title('Hyperbolic Tangent Function ')
plt.plot(x_axis_values,y_axis_values_tanh,color='black')


```
